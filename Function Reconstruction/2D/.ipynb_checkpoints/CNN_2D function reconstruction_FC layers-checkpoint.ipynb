# Generating function from equn 21 in StencilNet paper

```python
import math
import numpy as np
import torch
import matplotlib.pyplot as plt 
from tqdm import tqdm
import time
import numpy.random as npr
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

torch.set_default_tensor_type(torch.DoubleTensor)
```

```python
N = 20
L = 2*math.pi
x = np.linspace(0,10,200)
t = np.linspace(0,10,200)
f = 0.0
del_x = x[4]-x[3]
del_t = t[4]-t[3]
X,T = np.meshgrid(x, t)   #grid of point

np.random.seed(0)
for i in range(1,N+1):
    A = np.random.uniform(low = -0.1, high = 0.1)
    w = np.random.uniform(low = -0.4, high = 0.4)
    phi = np.random.uniform(low = 0.0, high = 2*math.pi)
    l = np.random.choice([2,3,4,5])
    f = f + A*np.sin(w*T + (2*math.pi*l*X)/L +phi)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, T, f, cmap='viridis', edgecolor='none')
ax.set_xlabel("x")
ax.set_ylabel("t")
ax.set_zlabel("f")

plt.show()
```

```python
f = torch.tensor(f)
X = torch.tensor(X)
T = torch.tensor(T)
del_x = torch.tensor(del_x)
del_t = torch.tensor(del_t)
```

```python
def preprocess(data):
    inputs = torch.DoubleTensor(data.size(0)-2,data.size(1)-2,4)
    outputs = torch.DoubleTensor(data.size(0)-2,data.size(1)-2,1)
    inputs = inputs.cuda()
    outputs = outputs.cuda()
    inputs = torch.autograd.Variable(inputs, requires_grad=True)
    outputs = torch.autograd.Variable(outputs, requires_grad=True)
    k = 0
    for i in range(1, data.size(0)-1):
        l = 0
        for j in range(1, data.size(1)-1): 
            inputs.data[k][l] = torch.DoubleTensor([data[i-1][j-1],data[i-1][j+1],data[i+1][j+1],data[i+1][j-1]])
            outputs.data[k][l] = data[i][j]
            l = l + 1
        k = k + 1
        
    return inputs,outputs

inputs,outputs = preprocess(f)
#print(f[0][0],f[0][1],f[0][2])
#print(f[1][0],f[1][1],f[1][2])
#print(f[2][0],f[2][1],f[2][2])
#print(inputs[0][0])
#print(outputs[0][0])
print(inputs.shape)
print(outputs.shape)
```

```python
N = inputs.size(0)
testPct = 0.5     #train-test split
nTrain = int(N*(1.-testPct))

train_inputs = inputs[:nTrain]
train_outputs = outputs[:nTrain]
test_inputs = inputs[nTrain:]
test_outputs = outputs[nTrain:]
print(train_inputs.shape)
print(test_inputs.shape)
print(train_outputs.shape)
print(test_outputs.shape)
```

```python
class Reconstruct(torch.nn.Module):
    def __init__(self,seed = 0):
        super().__init__()
        torch.manual_seed(seed)               #input dim is (99,198,4)  stencil length is 4
        self.fc1 = torch.nn.Conv1d(198,198,3)     #198 filters with filter size 3 each, output of this layer is (99,198,2) 
        self.fc2 = torch.nn.Linear(2,4)         #output of this layer is (99,198,4)
        
    def forward(self,x):
        x = self.fc1(x)
        x = torch.nn.functional.elu(self.fc2(x))
        return x
```

```python
def train(epoch,model,optimizer,train_inputs,train_outputs):
        
    optimizer.zero_grad()
    pred = model(train_inputs)
    loss = torch.nn.MSELoss()((pred*train_inputs).sum(2), train_outputs.flatten(1))
    loss.backward()
    optimizer.step()
    print('Epoch: {} \tLoss: {:.4f}'.format(epoch, loss))
    
    return loss.item()
    
def test(epoch,model,optimizer,test_inputs,test_outputs):
        
    pred = model(test_inputs)
    loss = torch.nn.MSELoss()((pred*test_inputs).sum(2), test_outputs.flatten(1))
        
    print('TEST SET RESULTS')
    print('Average loss: {:.4f}'.format(loss))
    
    return loss.item()
```

```python
seed = 1

np.random.seed(seed)
torch.manual_seed(seed)

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

model = Reconstruct(seed = seed)
if torch.cuda.is_available():
    model = model.cuda()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
test_loss = []
train_loss = []
epochs = 100
start = time.time()
for epoch in tqdm(range(epochs)):
    trainloss = train(epoch,model,optimizer,train_inputs,train_outputs)
    testloss = test(epoch,model,optimizer,test_inputs,test_outputs)
    test_loss.append(testloss)
    train_loss.append(trainloss)

print ("Training time : ",time.time()-start)
```

```python
plt.plot(test_loss,label = "Test Loss")
plt.plot(train_loss,label = "Train Loss")
plt.legend(loc = "upper right")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.title("Loss trajectories for network without moment constraints")
print("Training time is 8.8s")
print("Test loss after 100 epochs:",test_loss[-1])
print("Training loss after 100 epochs:",train_loss[-1])
```

```python
pred = model(test_inputs)

f_pred = (pred*test_inputs).sum(2)
f_act = test_outputs.flatten(1)
f_pred = f_pred.to('cpu').detach().numpy()
f_act = f_act.to('cpu').detach().numpy()


fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X[-100:-1,1:199],T[-100:-1,1:199],f_pred, cmap='viridis', edgecolor='none')
ax.set_title("Predicted function")
ax.set_xlabel("x")
ax.set_ylabel("t")
ax.set_zlabel("f")
plt.show()

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X[-100:-1,1:199],T[-100:-1,1:199],f_act, cmap='viridis', edgecolor='none')
ax.set_title("Original function")
ax.set_xlabel("x")
ax.set_ylabel("t")
ax.set_zlabel("f")
plt.show()
```
