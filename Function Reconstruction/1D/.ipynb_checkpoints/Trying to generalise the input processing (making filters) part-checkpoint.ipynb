# Generating function from equn 21 in StencilNet paper

```python
import math
import numpy as np
import torch
import matplotlib.pyplot as plt 
from tqdm import tqdm
import time
import numpy.random as npr

torch.set_default_tensor_type(torch.DoubleTensor)
```

```python
t = 0
N = 20
L = 2*math.pi
x = torch.linspace(0,10,200)
f = torch.tensor(0.0)

np.random.seed(0)
for i in range(1,N+1):
    A = np.random.uniform(low = -0.1, high = 0.1)
    w = np.random.uniform(low = -0.4, high = 0.4)
    phi = np.random.uniform(low = 0.0, high = 2*math.pi)
    l = np.random.choice([2,3,4,5])
    f = f + A*torch.sin(w*t + (2*math.pi*l*x)/L +phi)


f_noisy = f + npr.normal(0, 0.05, (200))
plt.plot(x,f,label = 'Original function')
plt.plot(x,f_noisy,label = 'Corrupted function with noise = 0.05')
plt.xlabel('x')
plt.ylabel('f(x,0)')
plt.legend()

N = f.size(0)
testPct = 0.5     #train-test split
nTrain = int(N*(1.-testPct))

train_data = f[:nTrain]
test_data = f[nTrain:]
train_data_noisy = f_noisy[:nTrain]
test_data_noisy = f_noisy[nTrain:]
```

```python
def preprocess(data):
    inputs = torch.DoubleTensor(data.size(0)-2,2)
    outputs = torch.DoubleTensor(data.size(0)-2,1)
    inputs = inputs.cuda()
    outputs = outputs.cuda()
    inputs = torch.autograd.Variable(inputs, requires_grad=True)
    outputs = torch.autograd.Variable(outputs, requires_grad=True)
    for i in range(0, data.size(0)-2):
        inputs.data[i] = torch.DoubleTensor([data[i],data[i+2]])
        outputs.data[i] = data[i+1]
        
    return inputs,outputs

#train_inputs,train_outputs = preprocess(train_data)
#train_inputs= train_inputs.view(train_inputs.size(0),2,1)
#test_inputs,test_outputs = preprocess(test_data)
#test_inputs= test_inputs.view(test_inputs.size(0),2,1)
#train_inputs_noisy,train_outputs_noisy = preprocess(train_data_noisy)
#test_inputs_noisy,test_outputs_noisy = preprocess(test_data_noisy)
#train_inputs_noisy = train_inputs_noisy.view(train_inputs_noisy.size(0),2,1)
#test_inputs_noisy = test_inputs_noisy.view(test_inputs_noisy.size(0),2,1)
```

```python
class Reconstruct(torch.nn.Module):
    def __init__(self,stencilSize,seed = 0):
        super().__init__()
        torch.manual_seed(seed)
        self.stencilSize = stencilSize
        self.fc1=torch.nn.Conv1d(2,16,1)
        self.fc2=torch.nn.Conv1d(16,2,1)
        
    def forward(self,x):
        x = self.preprocess(x)
        px = x.clone()
        x = self.fc1(x)
        x = torch.nn.functional.elu(self.fc2(x))
        return (x*px).sum(1)
    
    def preprocess(self,x) :
        inputs = torch.DoubleTensor(x.size(0)-self.stencilSize,self.stencilSize)
        inputs = inputs.cuda()
        inputs = torch.autograd.Variable(inputs, requires_grad=True)
        for i in range(0, x.size(0)-self.stencilSize):
            temp = []
            for j in range(i,i+self.stencilSize+1):
                if (j-i) != int(self.stencilSize/2):
                    temp.append(x[j])
            inputs.data[i] = torch.DoubleTensor(temp)
         
        inputs = inputs.view(inputs.size(0),2,1)
        return inputs
```

```python
def train(epoch,model,optimizer,train_data,train_outputs,stencilSize):
    
    optimizer.zero_grad()
    pred = model(train_data)
    loss = torch.nn.MSELoss()(pred, train_outputs)
    loss.backward()
    optimizer.step()
    print('Epoch: {} \tLoss: {:.4f}'.format(epoch, loss))
    
    return loss.item()
    
def test(epoch,model,optimizer,test_data,test_outputs,stencilSize):
    
    pred = model(test_data)
    loss = torch.nn.MSELoss()(pred, test_outputs)
        
    print('TEST SET RESULTS')
    print('Average loss: {:.4f}'.format(loss))
    
    return loss.item()
```

```python
seed = 1

np.random.seed(seed)
torch.manual_seed(seed)

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

model = Reconstruct(seed = seed,stencilSize = 2)
if torch.cuda.is_available():
    model = model.cuda()
for param in model.parameters():
    print(param)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
def make_outputs(data, stencilSize):
    outputs = torch.DoubleTensor(data.size(0)-stencilSize,1)
    outputs = torch.autograd.Variable(outputs, requires_grad=True)
    outputs.data = data[int(stencilSize/2):int(-stencilSize/2)]
    outputs = outputs.cuda()
    outputs = outputs.view(outputs.size(0),1)
    return outputs    
        
    
train_outputs = make_outputs(train_data,2)
test_outputs = make_outputs(test_data,2)
test_loss = []
train_loss = []
epochs = 100
start = time.time()
for epoch in tqdm(range(epochs)):
    trainloss = train(epoch,model,optimizer,train_data,train_outputs,2)
    testloss = test(epoch,model,optimizer,test_data,test_outputs,2)
    test_loss.append(testloss)
    train_loss.append(trainloss)

print ("Training time : ",time.time()-start)
```

```python
plt.plot(test_loss,label = "Test Loss")
plt.plot(train_loss,label = "Train Loss")
plt.legend(loc = "upper right")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.title("Loss trajectories for 2 FC layer with elu activation")
print("Training time is 2.96s")
print("Test loss after 100 epochs:",test_loss[-1])
print("Training loss after 100 epochs:",train_loss[-1])
```
